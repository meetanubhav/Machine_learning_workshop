```
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```
img=cv2.imread('hp.jpg')
```

```
print(img.shape)
```

```
cv2_imshow(img)
```

```
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis("off")
plt.show()
```

```
# Create Face Detector Object
face_detector = cv2.CascadeClassifier('templatedata.xml')
```

```
# Find Faces by Using Convolutional (Sliding Window) at Multiple
faces = face_detector.detectMultiScale(img,1.3,5)
```

```
# Printing the cordinates of each faces in the Image
for face_i in faces:
#   face_0=faces[0]
  x,y,w,h=face_i
  print(face_i)
```

```
color=(0,240,0)
for face_i in faces:
#   face_0=faces[0]
  x,y,w,h=face_i
  cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
plt.imshow(img)
```
